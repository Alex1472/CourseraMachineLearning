```python
import math
import numpy as np
from matplotlib import pylab as plt
from scipy import optimize

def f(x):
    return math.sin(x / 5.0) * math.exp(x / 10.0) + 5 * math.exp(-x / 2.0)

def g(x):
    return int(math.sin(x / 5.0) * math.exp(x / 10.0) + 5 * math.exp(-x / 2.0))
    
x = np.array(np.arange(0, 10, 0.1))
y = np.sin(x / 5.0) * np.exp(x / 10.0) + 5 * np.exp(-x / 2.0)
plt.plot(x, y)
plt.show()

x = np.array(np.arange(0, 40, 0.1))
y = np.sin(x / 5.0) * np.exp(x / 10.0) + 5 * np.exp(-x / 2.0)
plt.plot(x, y)
plt.show()

# Problem 1
print optimize.minimize(f, 2, method='BFGS')
print "\n"
print optimize.minimize(f, 30, method='BFGS')
print "\n"

# Problem 2
print optimize.differential_evolution(f, [(1, 30)])

#Problem 3
y = [g(i) for i in x]
plt.plot(x, y)
plt.show()

x = np.array(np.arange(0, 10, 0.1))
y = [g(i) for i in x]
plt.plot(x, y)
plt.show()

print optimize.minimize(g, 30, method='BFGS')
print "\n"
print optimize.differential_evolution(g, [(1, 30)])


```
