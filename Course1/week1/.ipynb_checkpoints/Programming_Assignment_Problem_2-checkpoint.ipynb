```python
import numpy as np
from matplotlib import pylab as plt

x = np.arange(1, 16, 0.1)
y = np.sin(x / 5.0) * np.exp(x / 10.0) + 5 * np.exp(- x / 2.0)

plt.plot(x, y)
plt.show()

x = np.array([1, 4, 10, 15])
a = []
for i in x:
    tmp = []
    for j in range(4):
        a.append(i ** j)

a = np.array(a)
a = a.reshape(4, 4)
b = np.sin(x / 5.0) * np.exp(x / 10.0) + 5 * np.exp(- x / 2.0)
res = np.linalg.solve(a, b)
x = np.arange(1, 16, 0.1)
y = res[0] + x * res[1] + x ** 2 * res[2] + x ** 3 * res[3] 
plt.plot(x, y)
plt.show()

print res
```
